```python
import pandas as pd
import os
from uniformplanets import PACKAGEDIR
import matplotlib.pyplot as plt
import numpy as np
import lightkurve as lk
from lightkurve import MPLSTYLE
```

```python
dirs = !ls ../outputs/
```

```python
rs = []
rp = []
per = []
rs_err = []
rp_err = []
per_err = []
lit_rs = []
lit_rp = []
lit_per = []
lit_rs_err = []
lit_rp_err = []
lit_per_err = []

datadir = os.path.abspath(os.path.join(PACKAGEDIR, os.pardir))
data = pd.read_csv(os.path.join(datadir, 'data', 'planets_2019.04.02_11.43.16.csv'), skiprows=range(81))

for d in dirs:
    dpath = os.path.join(datadir, 'outputs', d)
    try:
        # read and store published values
        lit_r_star = data[data['pl_hostname']==d][data['pl_letter']=='b']['st_rad'].values[0]
        lit_r_planet = data[data['pl_hostname'] == d][data['pl_letter']=='b']['pl_radj'].values[0]
        lit_period = data[data['pl_hostname'] == d][data['pl_letter']=='b']['pl_orbper'].values[0]
        lit_r_star_err = data[data['pl_hostname'] == d][data['pl_letter']=='b']['st_raderr1'].values[0]
        lit_r_planet_err = data[data['pl_hostname'] == d][data['pl_letter']=='b']['pl_radjerr1'].values[0]
        lit_period_err = data[data['pl_hostname'] == d][data['pl_letter']=='b']['pl_orbpererr1'].values[0]
        lit_rs.append(lit_r_star)
        lit_rp.append(lit_r_planet)
        lit_per.append(lit_period)
        lit_rs_err.append(lit_r_star_err)
        lit_rp_err.append(lit_r_planet_err)
        lit_per_err.append(lit_period_err)
    
        # read and store my values
        params = pd.read_csv(os.path.join(dpath, d + '_fit_parameters.csv'))
        r_star = params[params['Unnamed: 0'] == 'r_star']['mean'].values[0]
        r_star_err = params[params['Unnamed: 0'] == 'r_star']['sd'].values[0]
        r_planet = params[params['Unnamed: 0'] == 'r_pl__0']['mean'].values[0]
        r_planet_err = params[params['Unnamed: 0'] == 'r_pl__0']['sd'].values[0]
        period = params[params['Unnamed: 0'] == 'period__0']['mean'].values[0]
        period_err = params[params['Unnamed: 0'] == 'period__0']['sd'].values[0]
        rs.append(r_star)
        rp.append(r_planet)
        per.append(period)
        rs_err.append(r_star_err)
        rp_err.append(r_planet_err)
        per_err.append(period_err)
    except:
        continue
    try:
        # second planet
        lit_r_planet = data[data['pl_hostname'] == d][data['pl_letter']=='c']['pl_radj'].values[0]
        lit_r_planet_err = data[data['pl_hostname'] == d][data['pl_letter']=='c']['pl_radjerr1'].values[0]
        lit_rs.append(lit_r_star)
        lit_rp.append(lit_r_planet)
        lit_rs_err.append(lit_r_star_err)
        lit_rp_err.append(lit_r_planet_err)
        
        lit_period = data[data['pl_hostname'] == d][data['pl_letter']=='c']['pl_orbper'].values[0]
        lit_period_err = data[data['pl_hostname'] == d][data['pl_letter']=='c']['pl_orbpererr1'].values[0]
        lit_per.append(lit_period)
        lit_per_err.append(lit_period_err)
        period = params[params['Unnamed: 0'] == 'period__1']['mean'].values[0]
        period_err = params[params['Unnamed: 0'] == 'period__1']['sd'].values[0]
        per.append(period)
        per_err.append(period_err)

        r_planet = params[params['Unnamed: 0'] == 'r_pl__1']['mean'].values[0]
        r_planet_err = params[params['Unnamed: 0'] == 'r_pl__1']['sd'].values[0]
        rs.append(r_star)
        rp.append(r_planet)
        rs_err.append(r_star_err)
        rp_err.append(r_planet_err)
    except:
        continue
    try:
        # 3rd planet
        lit_r_planet = data[data['pl_hostname'] == d][data['pl_letter']=='d']['pl_radj'].values[0]
        lit_r_planet_err = data[data['pl_hostname'] == d][data['pl_letter']=='d']['pl_radjerr1'].values[0]
        lit_rs.append(lit_r_star)
        lit_rp.append(lit_r_planet)
        lit_rs_err.append(lit_r_star_err)
        lit_rp_err.append(lit_r_planet_err)
        
        lit_period = data[data['pl_hostname'] == d][data['pl_letter']=='d']['pl_orbper'].values[0]
        lit_period_err = data[data['pl_hostname'] == d][data['pl_letter']=='d']['pl_orbpererr1'].values[0]
        lit_per.append(lit_period)
        lit_per_err.append(lit_period_err)
        period = params[params['Unnamed: 0'] == 'period__2']['mean'].values[0]
        period_err = params[params['Unnamed: 0'] == 'period__2']['sd'].values[0]
        per.append(period)
        per_err.append(period_err)
        
        r_planet = params[params['Unnamed: 0'] == 'r_pl__2']['mean'].values[0]
        r_planet_err = params[params['Unnamed: 0'] == 'r_pl__2']['sd'].values[0]
        rs.append(r_star)
        rp.append(r_planet)
        rs_err.append(r_star_err)
        rp_err.append(r_planet_err)
    except:
        continue
```

```python
with plt.style.context(MPLSTYLE):
    # fig = plt.figure(figsize=(12, 7))
    # plt.errorbar(lit_rs, lit_rp, xerr=lit_rs_err, yerr=lit_rp_err, fmt='ro', label='Published Values')
    # plt.errorbar(rs, rp, xerr=rs_err, yerr=rp_err, fmt='ko', label='New Fit Parameters')
    plt.errorbar(lit_per, lit_rp, xerr=lit_per_err, yerr=lit_rp_err, fmt='ro', label='Published Values')
    plt.errorbar(per, [r*9.95520159283 for r in rp], xerr=rs_err, yerr=per_err, fmt='ko', label='New Fit Parameters')
    plt.xlabel('Orbital Period (days)')
    plt.ylabel(r'Planet Radius (R$_{Jup}$)')
    plt.yscale('log')
    plt.legend(loc=0)
```

```python
with plt.style.context(MPLSTYLE):
    plt.plot(np.linspace(-1,100,100), np.linspace(-1,100,100), 'r--')
    plt.errorbar(lit_per[:len(per)], per, xerr=lit_per_err[:len(per)], yerr=per_err, fmt='ko')
    plt.xlabel('Published Period (days)')
    plt.ylabel('Fit Period (days)')
    plt.xlim([0, 45])
    plt.ylim([0, 45])
```
